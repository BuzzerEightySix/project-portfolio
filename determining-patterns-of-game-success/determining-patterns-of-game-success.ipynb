## Изучение общих данных

Импортируем необходимые библиотеки и прочитаем файл датафрейма. Выведем первые 5 строк и дополнительно запросим общую информацию.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats as st

df = pd.read_csv('/datasets/games.csv')
display(df.head())
df.info()
```

Датасет содержит следующие столбцы:

- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий)

- `JP_sales` — продажи в Японии (миллионы проданных копий)
- `Other_sales` — продажи в других странах (миллионы проданных копий)
- `Critic_Score` — оценка критиков (максимум 100)
- `User_Score` — оценка пользователей (максимум 10)
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Как видно, не во всех столбцах количество значений совпадает с числом строк - явно имеются пропуски, но отсутствуют явные дубликаты.

Кроме того, нужно привести названия столбцов и их содержимое к нижнему регистру.

## Подготовка данных

### Замена названий столбцов

Приведем названия столбцов и содержимое столбцов Name, Platform и Genre к нижнему регистру с помощью str.lower()

```python
df.columns = df.columns.str.lower()

text_columns = ['name', 'platform', 'genre'] #в переменной сохраним названия столбцов, 
                                             #а далее в цикле применим функцию str.lower()

for column in text_columns:
    df[column] = df[column].str.lower()
df.columns   #выведем названия столбцов для проверки
```

### Преобразование типов данных

Изменим тип данный в столбце year_of_release на целочисленный тип int, предварительно удалив пропуски, так как в датасете их менее 2%.

```python
df = df.dropna(subset=['year_of_release'])
df['year_of_release'] = df['year_of_release'].astype(int)
```

В столбце user_score нужно перевести данные к типу float. Tbd в этом столбце означает to be determined - "еще не определено". Можно эти значения считать пропусками.

```python
df.loc[df['user_score'] == 'tbd', 'user_score'] = float('nan')
df['user_score'] = df['user_score'].astype(float)
```

### Обработка пропусков

Пропуски в столбце year_of_release и genre заполнять не будем, так как позже мы их всё равно отфильтруем, к тому же их довольно мало. Пропуски в столбцах critic_score и user_score тоже заполнять не будем, как отмечалось выше, из-за того что заполнение исказит результаты дальнейшего анализа и статистических тестов. В столбце user_score пропуски удалим позже, в ходе проверки гипотез.

Пропуски в rating заполним значением "без рейтинга", т.к. это категориальная переменная.

```python
df['rating'] = df['rating'].fillna('без рейтинга')
df['rating'].unique() #проверим уникальные значения в столбце
```

Есть 2 пропущенных значения в столбцах name и genre. Так как количество незначительное - удалим эти пропуски.

```python
df = df.dropna(subset=['name', 'genre'])
df = df.reset_index(drop=True) #сбросим индексы после удаления пустых значений
```

### Подсчет суммарных продаж на всей территории

Создадим столбец total_sales, в котором просуммируем продажи во всех регионах.

```python
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
```

### Вывод

В ходе подготовки данных для анализа мы привели названия столбцов и содержимое текстовых столбцов к нижнему регистру.

Значения tbd в столбце user_score мы приравняли к пропускам и изменили тип данных в этом столбце. Преобразовывать типы данных в других столбцах мы не стали, чтобы не искажать результаты дальнейшего анализа данных.

По этой же причине мы отказались от заполнения пропусков в столбцах critic_score и user_score. В столбцах year_of_release, genre и name пропуски были удалены, а в столбце year_of_release тип данных изменен на целочисленный. Пропуски в столбце rating заполнили "заглушкой" "без рейтинга".

Пропуски в столбцах с данными об оценках пользователей (user_score) и оценках критиков (critic_score) можно объяснить тем, что на момент формирования датафрейма соответствующие общие оценки не были сформированы либо из-за отсутствия отдельных оценок, либо из-за их недостаточного количества этих оценок для вывода общей оценки для игры. В столбце с рейтингом от ESRB тоже можно объяснить тем, что рейтинг игре не присваивался. Пропуски в столбцах с названием игры (name) и годом выпуска (year_of_release) можно объяснить тем, что данные могли собираться из разных источников и в отдельных источниках мультиплатформенные игры могли быть записаны по-разному. Где-то могла быть общая дата на всех платформах, где-то разные. На каких-то платформах могло отличаться название, в связи с чем из-за разницы метода записи данных в разных источниках появились пропуски.

Помимо прочего, были посчитаны суммарные продажи во всех регионах и записаны в отдельный столбец total_sales.

## Исследовательский анализ данных

Проверим сколько игр выпускалось в разные годы.

```python
df['year_of_release'].hist(bins=37, figsize=(10, 5))
plt.suptitle('Количество выпущенных игр в разные годы')
plt.xlabel('годы')
plt.ylabel('кол-во игр')
plt.xticks(np.arange(1980, 2017))
plt.xticks(rotation = 90)
plt.show()
```

Как видно, мы имеем данные с 1980 года. Количество выпущенных игр в разные периоды сильно отличается, поэтому производить анализ по всей выборке будет неправильным. В разные периоды появлялись и исчезали разные платформы и, можно предположить, что с этим фактом связана такая большая разница в количестве игр, выпущенных в разные годы. В связи с этим нам необходимо понять сколько в среднем "живет" платформа и на основе этой информации выделить актуальный период, с которым и будем работать дальше.

Посмотрим, как менялись продажи по платформам. Выберем 10 платформ с наибольшими суммарными продажами и построим график распределения по годам.

```python
platform_sales = df.groupby('platform').agg({'total_sales': 'sum'}).sort_values(
    by='total_sales', ascending=False).head(10).index
df.query('platform in @platform_sales').pivot_table(
    index='year_of_release', columns='platform', values='total_sales', aggfunc='sum').plot(
    figsize=(10, 5), grid=True)
plt.suptitle('Продажи по 10 самым популярным платформам')
plt.xlabel('годы')
plt.ylabel('проданные игры, млн. копий')
plt.show()
```

На первый взгляд, судя по графику, исчезают старые и появляются новые платформы в течение 7-8 лет. Судя по графику потенциально самыми прибыльные платформы - ps с середины 90-х до середины нулевых, ps2 - с начала нулевых до конца десятых (самая прибыльная платформа), wii, ps3 и x360 с середины нулевых и до 2016 года. Очевидно что жизненный цикл этих платформ подошел к концу и самые прибыльные платформы в будущем на этом графике отсутствуют. Тем не менее, жизненный цикл платформы ps4 в самом начале и, вероятно, пик продаж этой платформы ещё впереди. Можно считать что это одна из перспективных платформ.

Далее найдем актуальный период, по которому будем вести дальнейший анализ. Для этого произведем группировку по столбцам platform и year_of_release и возьмем срез с продажами более 1 млн. в качестве ориентира. Сохраним результаты группировки в переменную actual_period. далее рассчитаем жизненный период каждой платформы, произведя ещё одну группировку, применяя функцию count к столбцу year_of_release. Ну и далее уже найдем среднее значение по получившемуся столбцу.

```python
actual_period = df.groupby(['platform','year_of_release']).agg(
    {'total_sales': 'sum'}).query('total_sales > 1')
actual_period = actual_period.reset_index().groupby('platform').agg(
    {'year_of_release': 'count'}) # расчитаем жизненный период каждой платформы
actual_period.mean() # расчитает сколько в среднем "живет" платформа
```

В среднем жизненный цикл платформы составляет 7.92 года, но по графику видно, что продажи игр по отдельным платформам накладываются одни на другие, из-за чего можно сделать вывод, что в определенный промежуток, ориентировочно в 3-4 года, игры продаются для предыдущих поколений платформ и для новых. Поэтому лучше сузить актуальный период до 3 лет, чтобы учитывать максимально актуальные платформы и не учитывать платформы, срок жизни которых подходит к концу. Так как нам надо спрогнозировать данные за 2017 год - актуальным периодом будут продажи с 2014 года по 2016. Сделаем срез данных по этому периоду и сохраним полученный датафрейм в переменной actual_df. Все дальнейшие действия будем производить именно с ним.

```python
actual_df = df.query('year_of_release >= 2014').reset_index()
```

Построим график «ящик с усами» по глобальным продажам игр в разбивке по платформам.

```python
actual_df.boxplot(column='total_sales', by='platform', figsize=(10, 5))
plt.ylim(0, 1.60)
plt.suptitle('Диаграмма размаха глобальных продаж с разбивкой по платформам')
plt.show()
```

По диаграмме видно, что медианные продажи ps4 и xone самые высокие, что свидетельствует об успешности этих платформ в своём поколении. От них отстают платформы ps3 и x360, чей жизненный цикл уже подходит к концу, но которые ранее входили в топ-10 самых популярных платформ. Помимо этих консолей высокие медианные продажи у платформ wii и wiiu, но первая находится внизу топа-10, а wiiu не входит в топ-10 в принципе.

Посмотрим, как влияют на продажи платформ отзывы пользователей и критиков. Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами. Для платформ psp и wii расчеты делать не будем, так как выборка по этим платформам нерепрезентативна.

Сначала рассмотрим данные по отзывам пользователей.

```python
top_platforms = actual_df.query('platform != ["psp", "wii"]')['platform'].unique() #топ платформ, входящих в актуальный период

for i in top_platforms:
    score_sales_games = actual_df.query('platform==@i')
    score_sales_games.plot(x='user_score', y='total_sales', kind='scatter')
    plt.suptitle(f'Диаграмма зависимости продаж от оценок пользоввателей по платформе {i}')
    print(f'Коэффиент корреляции между отзывами и продажами по платформе {i}:',
          score_sales_games['total_sales'].corr(score_sales_games['user_score']))
```

Корреляция очень слабая и не везде одинаковая, если смотреть графики: по перспективным платформам ps4, xone можно заметить что чем оценка выше - тем продаж больше. Коэффициенты корреляции указывают на присутствие взаимосвязи только по данным платформы wiiu, что может нам говорить о нелинейной зависимости.

Аналогичные манипуляции проведем с данными по отзывам критиков. И придем к промежуточному выводу.

```python
for i in top_platforms:
    score_sales_games = actual_df.query('platform==@i')
    score_sales_games.plot(x='critic_score', y='total_sales', kind='scatter')
    plt.suptitle(f'Диаграмма зависимости продаж от оценок критиков по платформе {i}')
    print(f'Коэффиент корреляции между отзывами и продажами по платформе {i}:',
          score_sales_games['total_sales'].corr(score_sales_games['critic_score']))
```

По диаграммам видно, что зависимость продаж от оценок пользователей и оценок критиков присутствует. Распределение точек более плотное в районе высоких оценок, что говорит о том, что, как правило, чем у игры выше оценка - тем выше у неё продажи. Не на всех графиках распределение одинаковое, что говорит нам о том, что не все платформы полно представлены в нашей выборке, что может быть связано с окончанием их жизненного цикла. 

Однако, если посмотреть на коэффициенты корреляции - можно подумать что зависимости нет, либо она очень слабая. Особенно сильно это видно в расчетах коэффициентов по оценкам пользователей. Отсюда можно сделать вывод, что у зависимость нелинейная, более сложная, поэтому коэффициент корреляции на данном этапе исследования можно не учитывать. В случае с расчетами по оценкам критиков - тут корреляция более явная и коффициенты расположены гораздо более полярно, что может говорить о том, что оценки критиков больше влияют на продаже, чем оценки пользователей.

## Проверка гипотез

Для проверки есть две гипотезы:

первая гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

вторая гипотеза: средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

### Проверка первой гипотезы

Сформулируем нулевую и альтернативные гипотезы:

Нулевая гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

Альтернативная гипотеза: средние пользовательские рейтинги платформ Xbox One и PC различаются.

Для начала нам необходимо удалить строки с пропущенными рейтингами, чтобы не искажать картину наблюдений.

```python
actual_df_score = actual_df.dropna(subset=['user_score'])
```

Далее в отдельных перменных сохраним данные по платформам Xbox One и PC. Дополнительно найдем стандартное отклонение и дисперсию полученных выборок, а так же среднее значение. Обозначим границы интервала практически всех значений выборок.

```python
xone_score = (actual_df_score[actual_df_score['platform'] == 'xone']['user_score'])
pc_score = (actual_df_score[actual_df_score['platform'] == 'pc']['user_score'])

scores = [xone_score, pc_score]

for score in scores:
    score_std = np.std(score) #находим стандартное отклонение
    print(f'Стандартное отклонение:', score_std)
    score_var = np.var(score) #находим дисперсию
    print(f'Дисперсия:', score_var)
    score_mean = score.mean() #находим средний пользовательский рейтинг
    print('Средний пользовательский рейтинг платформы', score_mean)
    bottom_line = round(score_mean - (3 * score_std), 1) #посчитаем нижнюю границу интервала
    top_line = round(score_mean + (3 * score_std), 1) #посчитаем верхнюю границу интервала
    print(f'Интервал: {bottom_line} - {top_line}')
```

В соответствии с "правилом трёх сигм", почти все значения выборки xone_score находятся в интервале от 2.4 до 10.7, а для платформы pc - в интервале от 2 до 11.3

Для проверки гипотезы используем метод рscipy.stats.ttest_ind().

```python
alpha = 0.05 #пороговое значение (уровень значимости) - ориентируемся на него, чтобы выяснить,
             #является ли результат статистически значимым

results = st.ttest_ind(xone_score, pc_score)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### Проверка второй гипотезы

Сформулируем нулевую и альтернативные гипотезы:

Нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports одинаковые;

Альтернативная гипотеза: средние пользовательские рейтинги жанров Action и Sports разные.

Сначала выделим определим рейтинги у жанров и сохраним соответствующие данные в отдельных переменных. Так же, как и ранее для платформ найдем стандартное отклонение и дисперсию полученных выборок, а так же среднее значение. Обозначим границы интервала практически всех значений выборок.

```python
action_score = (actual_df_score[actual_df_score['genre'] == 'action']['user_score'])
sports_score = (actual_df_score[actual_df_score['genre'] == 'sports']['user_score'])

scores = [action_score, sports_score]

for score in scores:
    score_std = np.std(score) #находим стандартное отклонение
    print(f'Стандартное отклонение:', score_std)
    score_var = np.var(score) #находим дисперсию
    print(f'Дисперсия:', score_var)
    score_mean = score.mean() #находим средний пользовательский рейтинг
    print('Средний пользовательский рейтинг платформы', score_mean)
    bottom_line = round(score_mean - (3 * score_std), 1) #посчитаем нижнюю границу интервала
    top_line = round(score_mean + (3 * score_std), 1) #посчитаем верхнюю границу интервала
    print(f'Интервал: {bottom_line} - {top_line}')
```

Ориентируясь на "правило трёх сигм" почти все значения выборки action_score находятся в интервале от 2.7 до 10.9, а для выборки sports_score - в интервале от 1 до 11.2.

Аналогичным образом используем рscipy.stats.ttest_ind().

```python
results = st.ttest_ind(action_score, sports_score)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### Пояснения к гипотезам

Нулевая гипотеза формулировалась исходя из того, что две совокупности, которые сравниваются по одному или нескольким признакам, равны. Альтернативная же гипотза формулировалась как противоположная по смыслу нулевой гипотезе.

Для проверки значимости использовался критерий t-критерий Стьюдента, так как мы проверяли равенство средних значений в двух выборках. Проверка осущесвтлялась с помощью t-теста (метод st.ttest_ind()), где потом с заранее заданным пороговым значением уровня значимости, мы сравнивали p-значение.

### Вывод

На данном этапе мы проверили 2 гипотезы и выявили что средний пользовательский рейтинг платформ Xbox one и PC одинаковый. Так же мы определелили, что средние пользовательские рейтинги жанров Action и Sports разные, опровергнув нулевую гипотезу о том, что рейтинги указанных жанров одинаковые. В данном случае сформулированная в условии гипотеза являлась альтернативной.
