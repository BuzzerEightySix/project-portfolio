## Предобработка данных

### Обработка пропущенных значений

Определим пропущенные значения в датафрейме с помощью методов isna().sum().

```python
df.isna().sum()
```

В столбце rooms имеются неявные пропуски, обозначенные нулевыми значениями. Для того чтобы выявить закономерность появления таких пропусков рассмотрим срез датафрейма со столбцами living_area, kitchen_area, studio и rooms, где значения rooms равны нулю.

```python
df.loc[:, ['living_area', 'kitchen_area', 'studio', 'rooms']].query('rooms == 0')
```

Можно сделать предположение, что нулевые значения в столбце rooms означает что указанные объекты недвижимости являются либо студиями, либо имеет место неполное предоставление данных пользвателями сервиса, которые указывали только общую площадь квартиры, но не указывали другие данные. Заменим пропуски значением 1 с помощью метода replace.

```python
df['rooms'] = df['rooms'].replace(0, 1)
df['rooms'].unique() #произведем проверку замены, просмотрев уникальные значения столбца
```

Проверим уникальные значения в столбце locality_name с помощью метода unique, так как там присутствуют типы населенных пунктов.

```python
print(df['locality_name'].unique())
```

Обнаружены дубликаты многих значений из-за того что по-разному указаны типы населенных пунктах в разных строках. Чтобы избавиться от дубликатов было принято решение убрать типы населенных пунктов из данных столбца с помощью цикла. Замена производится с помощью  str.replace(), будем перебирать все виды населенных пунктов и заменять их на пустое значение, оставив в данных только наименования населенных пунктов.

```python
for name in [
    'посёлок ',
    'городской ',
    'деревня ',
    'поселок ',
    'городского типа ',
    'садовое товарищество ',
    'коттеджный ',
    'при железнодорожной станции ',
    'садоводческое некоммерческое товарищество ',
    'село ',
    'станции '
]:
    df['locality_name'] = df['locality_name'].str.replace(name, '')

print(df['locality_name'].unique()) #для проверки выведем уникальные значения обновленного столбца locality_name
```

В столбце locality_name были обнаружены пропуски. Их совсем мало, менее 1%. Это категориальные значения и в данном случае заменить эти значения нечем. Принято решение удалить строки с этими пропущенными значениями. Для этого испольуем метод dropna(), в параметр subset передаем название столбца. В параметр inplace указываем True, который применит изменения в датафрейме.

Для проверок значений здесь и далее будут использованы isna().sum().

```python
print("Пропуски в столбце 'locality_name' до обработки:", df['locality_name'].isna().sum())
df.dropna(subset=['locality_name'], inplace=True)
print("Пропуски в столбце 'locality_name' после обработки:", df['locality_name'].isna().sum())
```

В столбце ceiling_height заполним пропущенные значения медианными значениями. Для этого используем loc без учета групп.

```python
print("Пропуски в столбце 'ceiling_height' до обработки:", df['ceiling_height'].isna().sum())
df.loc[df['ceiling_height'].isna(), 'ceiling_height'] = df['ceiling_height'].median()
print("Пропуски в столбце 'ceiling_height' после обработки:", df['ceiling_height'].isna().sum())
```

Так же медианными значениями заполним пропущенные значения в floors_total, но с группировкой по одной переменной. Для этого создадим цикл и в цикле среди уникальных значений столбца floor выявим медиану.

```python
print("Пропуски в столбце 'floors_total' до обработки:", df['floors_total'].isna().sum())
for floor in df['floor'].unique():
    median = df.loc[df['floor'] == floor, 'floors_total'].median()
    df.loc[(df['floors_total'].isna()) & (df['floor'] == floor), 'floors_total'] = median
print("Пропуски в столбце 'floors_total' после обработки:", df['floors_total'].isna().sum())
```

Аналогичным способом поступим с пропусками в столбце living_area.

```python
print("Пропуски в столбце 'living_area' до обработки:", df['living_area'].isna().sum())
for total_area in df['total_area'].unique():
    median = df.loc[df['total_area'] == total_area, 'living_area'].median()
    df.loc[(df['living_area'].isna()) & (df['total_area'] == total_area), 'living_area'] = median
print("Пропуски в столбце 'living_area' после первой обработки:", df['living_area'].isna().sum())
df.loc[df['living_area'].isna(), 'living_area'] = df['living_area'].median() #так как после первой обработки всё еще осталось
                                                                             #много пропусков - принято решение заполнить 
                                                                             #оставшиеся пропуски уже медианными значениями 
                                                                             #без учета групп, по аналогии с заменой пропущенных
                                                                             #значений в столбце ceiling_height 
print("Пропуски в столбце 'living_area' после второй обработки:", df['living_area'].isna().sum())
```

По аналогии с заполнением пропусков в living_area заполним пропуски в столбце kitchen_area.

```python
print("Пропуски в столбце 'kitchen_area' до обработки:", df['kitchen_area'].isna().sum())
for total_area in df['total_area'].unique():
    median = df.loc[df['total_area'] == total_area, 'kitchen_area'].median()
    df.loc[(df['kitchen_area'].isna()) & (df['total_area'] == total_area), 'kitchen_area'] = median
print("Пропуски в столбце 'kitchen_area' после первой обработки:", df['kitchen_area'].isna().sum())
df.loc[df['kitchen_area'].isna(), 'kitchen_area'] = df['kitchen_area'].median()
print("Пропуски в столбце 'kitchen_area' после второй обработки:", df['kitchen_area'].isna().sum())
```

Переименуем столбцы cityCenters_nearest, parks_around3000, ponds_around3000: приведем к "змеиному реестру" и уберем прописные буквы.

```python
df.rename(columns={'cityCenters_nearest': 'city_centers_nearest', 'parks_around3000': 'parks_around_3000',
                   'ponds_around3000': 'ponds_around_3000'}, inplace=True)
df.columns
```

Для заполнения пропусков в переменных airports_nearest, city_сenters_nearest, parks_around_3000, ponds_around_3000 объявим функцию fill_local_median, которая будет принимать данные столбца locality_name и столбца, где необходимо заполнить пропуски, а возвращать столбец с заполненными медианными значениями пропусками с  группировкой по столбцу locality_name.

```python
def fill_local_median(local, distance):
    print(f'Пропуски в столбце {distance} до обработки:', df[distance].isna().sum()) #в функцию встроим вывод числа пропусков до работы
                                                                         #функции
                                                                         #и после
    for value in df[local].unique():
        median = df.loc[df[local] == value, distance].median()
        df.loc[(df[distance].isna()) & (df[local] == value), distance] = median
        df.loc[df[distance].isna(), distance] = df[distance].median()
    print(f'Пропуски в столбце {distance} после обработки:', df[distance].isna().sum())
```

Для вышеуказанных столбцов применим функцию fill_local_median в цикле.

```python
columns = ['airports_nearest', 'city_centers_nearest', 'parks_around_3000', 'parks_nearest', 'ponds_around_3000', 'ponds_nearest']

for col_name in columns:
    fill_local_median('locality_name', col_name)
```

Есть пропуски в столбце is_apartment, содержащем булевы значения. Скорее всего эти значения пропущены из-за того что объекты недвижимости в этих строках не являются аппартаментами. Пользователи, создающие объявления обычно прямо указывают этот нюанс и логично предположить что отсутствие указания на то что объект является аппартаментами означает то, что объект аппартаментами не является. В связи с этим принято решение заполнить пропуски значениями False.

Применим к столбцу is_apartment метод fillna(), а качестве аргумента которого передадим False.

```python
print("Пропуски в столбце 'is_apartment' до обработки:", df['is_apartment'].isna().sum())
df['is_apartment'] = df['is_apartment'].fillna(False)
print("Пропуски в столбце 'is_apartment' после второй обработки:", df['is_apartment'].isna().sum())
```

Пропуски в колонке balcony скорее всего связаны с тем, что в соответствующих объектах балконы отсутствуют, так как их наличие явно было бы указано. В связи с этим заполним пропуски в этом столбце нулевыми значениями.

Используем аналогичный метод, как при заполнении пропусков в столбце is_apartment.

```python
print("Пропуски в столбце 'balcony' до обработки:", df['balcony'].isna().sum())
df['balcony'] = df['balcony'].fillna(0)
print("Пропуски в столбце 'balcony' после второй обработки:", df['balcony'].isna().sum())
```

### Изменение типов данных

Изменим типы данных в отдельных столбцах: в first_day_exposition - изменим тип данных на datetime, но отбросив время, в floors_total с float на int, в balcony с float на int.

Применим к floors_total, balcony метод astype() и передадим в качестве аргумента нужный нам тип данных. Сделаем это в цикле.

```python
columns = ['floors_total', 'balcony', 'parks_around_3000', 'ponds_around_3000']

for i in columns:
    df[i] = df[i].astype('int')
```

Для изменения типа данных в столбце first_day_exposition применим метод pandas to_datetime(), которому передадим столбец first_day_exposition и укажем формат в соответствии с записью данных в столбце, но отбросим из него данные по времени. Отбросим потому что столбцец содержит только дату, данные о времени там нулевые.

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT')
```

### Цена квадратного метра

Посчитаем цену квадратного метра. Для этого разделим данные столбца last_price на данные столбца total_area, сохранив значения в новом столбце square_metre_price. С помощью метода round() округлим полученные значения до 1 знака после точки.

```python
df['square_metre_price'] = (df['last_price'] / df['total_area']).round(1)
df['square_metre_price'].head() #для проверки вызовем 5 первых значений столбца
```

### День недели, месяц и год публикации объявления

Найдём и добавим в таблицу день недели, месяц и год публикации объявления.

Создадим новый столбец week_day, который будет содердать данные о дне недели. Применим к столбцу first_day_exposition метод dt.weekday и сохраним данные в созданный столбец.

```python
df['week_day'] = df['first_day_exposition'].dt.weekday
df['week_day'].value_counts() #посчитаем число уникальных значений столбца для того чтобы выяснить,
                              #какое значение соответствует конкретному дню недели
```

Объявим функцию weekday(), которая будет принимать уникальные значения из столбца week_day и возвращать привычные нам наименования дней недели. Применим функцию к столбцу week_day с помощью метода apply и сохраним полученные данные в новом столбце week_day_name.

```python
def weekday(day):
    if day == 0:
        return 'Понедельник'
    elif day == 1:
        return 'Вторник'
    elif day == 2:
        return 'Среда'
    elif day == 3:
        return 'Четверг'
    elif day == 4:
        return 'Пятница'
    elif day == 5:
        return 'Суббота'
    else:
        return 'Воскресенье'
    
df['week_day_name'] = df['week_day'].apply(weekday)

df.loc[:, ['week_day', 'week_day_name']].head(10) #проверим преобразования, отобразив первые 10 строк столбцов week_day и
                                                  #week_day_name
```

Создадим новый столбцец month, которому перерадим значения столбца first_day_exposition, к которому применим метод dt.month.

```python
df['month'] = df['first_day_exposition'].dt.month
df['month'].value_counts() #посчитаем число уникальных значений столбца для того чтобы выяснить,
                           #какое значение соответствует конкретному месяцу
```

Аналогичным образом, как в случае со столбцом weekday, создадим функцию month, с помощью которой преобразуем значения столбца month в названия месяцев и сохраним их в столбце month_name.

```python
def month(value):
    if value == 1:
        return 'Январь'
    elif value == 2:
        return 'Февраль'
    elif value == 3:
        return 'Март'
    elif value == 4:
        return 'Апрель'
    elif value == 5:
        return 'Май'
    elif value == 6:
        return 'Июнь'
    elif value == 7:
        return 'Июль'
    elif value == 8:
        return 'Август'
    elif value == 9:
        return 'Сентябрь'
    elif value == 10:
        return 'Октябрь'
    elif value == 11:
        return 'Ноябрь'
    else:
        return 'Декабрь'
    
df['month_name'] = df['month'].apply(month)

df.loc[:, ['month', 'month_name']].head(10) #проверим преобразования, отобразив первые 10 строк столбцов month и monthname
```

Создадим новый столбцец year, которому перерадим значения столбца first_day_exposition, к которому применим метод dt.year.

### Этаж квартиры

Сгруппируем этажи по категориям: первый, последний и другой.

Для этого объявим функцию floor_cat, которой передадим строки. Обозначим условия: если в строке значение этажа объекта равно единице - функция возвратит строку 'первый', если в строке значение этажа равно строкам со значениями общего числа этажей в здании - функция возвращает строку 'последний'. В остальных случаях функция вернет значение 'другой'.

Применим эту функицю к датафрейму с помощью apply и сохраним полученные значения в новом столбце floor_cat.

```python
def floor_cat(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
    
df['floor_cat'] = df.apply(floor_cat, axis=1)
df.loc[:, ['floor', 'floor_cat']].head(15) #проверим преобразования, отобразив первые 15 строк столбцов floor и floor_cat
```

### Cоотношение жилой и общей площади, а также отношение площади кухни к общей

Найдем соотношение жилой и общей площади. Для этого разделим значения столбца living_area на значение столбца total_area. Отобразим значения в процентах, для чего умножим получившиеся преобразования на 100 и округлим с помощью метода round до 1 знака после запятой. Сохраним полученный результат в новом столбце living_total_area_ratio.

```python
df['living_total_area_ratio'] = ((df['living_area'] / df['total_area']) * 100).round(1)
df['living_total_area_ratio'].head() #проверим преобразования, отобразив первые 5 значений столбца
```

Аналогичным образом произведем преобразования в столбце kitchen_area и сохраним новые полученные значения в столбце kitchen_total_area_ratio.

```python
df['kitchen_total_area_ratio'] = ((df['kitchen_area'] / df['total_area']) * 100).round(1)
df['kitchen_total_area_ratio'].head()
```

### Анализ десяти населённых пунктов с наибольшим числом объявлений

Нужно выявить 10 населенных пунктов с наибольшим числом объявлений. Для этого сгруппируем населенные пункты с посощью groupby, посчитаем для каждого число объявлений, отсортируем и выберем первые 10.

```python
local_top_10 = df_clean.groupby('locality_name')['is_apartment'].count().sort_values(ascending=False).head(10).index.values
#группируем по locality_name, на вывод можно взять любой столбец, например is_apartment
#с помощью index.values получим индексы
print(local_top_10)
```

Топ 10 нашли городов нашли. Отфильтруем датафрейм по этому списку и сохраним в новую переменную df_local_top_10.

```python
df_local_top_10 = df_clean[df_clean['locality_name'].isin(local_top_10)]
```

Посчитаем среднюю цену квадратного метра в этих населённых пунктах. Для этого построим сводную таблицу с групировкой датайфрейма df_local_top_10 по наименованию населенного пункта, в качестве значений таблицы выведем цену за квадратный метр. К этим значениям применим mean и count, которые выведут средние значения цены, а с помощью sort_values отсортируем таблицу по count.

```python
df_local_top_10.pivot_table(
    index='locality_name',
    values='square_metre_price',
    aggfunc=('mean', 'count')
).sort_values(
    by='count',
    ascending=False
)
```

Лидер по количеству объявлений Санкт-Петербург с очень большим отрывом от остальных городов рейтинга.

Создадим аналогичную сводную таблицу, только уже в целом для датафрейма df_clean, чтобы выявить самую высокую и самую низвкую среднюю цену за квадратный метр квартир. В новой таблице сортировку сделаем по столбцу square_metre_price, а из агрегатора функций внутри pivot_table уберем функцию count. Сохраним полученный датафрейм в переменную df_clean_mean_price и вытащим из него значения head(1) и tail(1).

```python
df_clean_mean_price = df_clean.pivot_table(
    index='locality_name',
    values='square_metre_price',
    aggfunc=('mean')
).sort_values(
    by='square_metre_price',
    ascending=False
)

display(df_clean_mean_price.head(1))
display(df_clean_mean_price.tail(1))
```

Выяснили, что самая высокая средняя цена квадратного метра жилья в Лисьем Носе, а самая низкая - в Серебрянском.
